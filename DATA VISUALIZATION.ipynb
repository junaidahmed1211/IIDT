# BAR CHART

# seaborn

# scatter plot 

we can use the .scatterplot method for creating a scatter plot,and just as in pandas we need to pass  it the coloumn names of the x and y

```python
sns.scatterplot(x='Sepal.Length',y='Sepal.Width',hue='Species',data=iris)

#we can also highlight the points by class during the huge argument
     #which is a lot easier than in matplotlib
#order for the levels of the huge variable in the plateet
    #set of colors for mapping the hue variable
```

# line chart
 to create a line-chart the sns.lineplot method can be used.the only required argument is the data,which in our case are the four numeric coloumns from the iris dataset

```python
sns.lineplot(data=iris.drop(['Species'],axis=1))
```

# histogram

to create a histogram in seaborn we use the sns.displot method. we need to pass it the coloumn we want to plot and it will calculate the occurences itself.we can also pass it the number of bins

```python
sns.distplot(wine['points'],bins=10,kde=False)
#kde:estimates the probability density function
```

```python
sns.distplot(wine['points'],bins=10,kde=True)
```

# BAR CHART

in seaborn a bar chart can be created using the sns.countplot method and passing it the data 

```python
sns.countplot(wine['points'])
```

# BOX PLOT
 A BOXPLOT  is a graphical method of displaying the five number summary we can create box plots using seaborns sns.boxplot method and passing it the data as well as the x and y coloumn name 

```python
df=wine[(wine['points']>=95) & (wine['price']<1000)]
sns.boxplot('points','price',data=df)
#vertical boxplot grouped by a variables:
```

# heat map

a Heat map is a graphical representatio of data where the individual values contained in a mtarix are represented as colors.heat maps are perfect for exploring the correlation of features in a data set   

```python
fig,ax=plt.subplots(figsize=(10,10))
sns.heatmap(iris.corr(),annot=True,square=True)
#annot: if tru,write the data value in each cell
```
