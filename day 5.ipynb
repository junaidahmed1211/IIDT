```python
mydict=[{"p":2,"q":3,"r":4,"s":5},{"p":20,"q":30,"r":40,"s":50},{"p":200,"q":300,"r":400,"s":500}]
x=pd.DataFrame(mydict)
x
```

```python
#2.b.
#indexing the rows with the list of integrs using loc/iloc

print(x.iloc[0,1])
print(x.iloc[:,3])
print(x.iloc[:,1:3])
```
