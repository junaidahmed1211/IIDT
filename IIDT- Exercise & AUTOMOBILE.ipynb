```python
#import packages
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

```python
#Create a DataFrame df from this dictionary data which has the index labels.
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}
labels =['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df=pd.DataFrame(data,index=labels)
df
```

```python
#Display a summary of the basic information about this DataFrame and its data.
df.describe()
```

```python
#Return the first 3 rows of the DataFrame df.
df.head(3)
```

```python
#Select just the 'animal' and 'age' columns from the DataFrame.
df[['animal','age']]
```

```python
#Select the data in rows [3, 4, 8] and in columns ['animal', 'age'].
df.iloc[[3,4,8]][['animal','age']]
```

```python
#Select the data in rows [3, 4, 8] and in columns ['animal', 'age'].
#df.loc[[3,4,8],['animal','age']]  
df.loc[['d','e','i'],['animal','age']]    #for loc we have to use only by index names as we defined in data frame
#iloc ---we can define row using as its position....
#iloc---if the lables are in categorical values we cant use it...if they are numeric we can define the rows
```

```python
#Select only the rows where the number of visits is greater than 2.
df.query('visits>2')
```

```python
#Select the rows where the age is missing, i.e. is NaN.
df[df['age'].isnull()]
```

```python
#alternate method
df.query('age=="NaN"')
```

```python
#count the missing values in the age variable
df.isnull().sum()
```

```python
#alterbate method
df['age'].isnull().sum()
```

```python
#Select the rows where the animal is a cat and the age is less than 3.
df.query('animal=="cat" and age<3')
```

```python
df[(df['animal']=="cat") &(df['age']<3)]    #in indexation method "and" will not be executed...there u should use "&"
```

```python
##Select the rows the age is between 2 and 4 (inclusive).
df.query('age>2&age<4')
```

```python
#alternsate method
df[df['age'].between(2,4)]
```

```python
#Change the age in row 'f' to 1.5.
df.loc[['f'],['age']]=1.5
df
```

```python
df['age']['f']=2.8
df
```

```python
#calulate the sum of all vists(the total number o f vists)
df['visits'].sum()
#df.visits.sum()
```

```python
#Calculate the mean age for each different animal in df.
df.groupby('animal').agg({'age':"mean"}).reset_index()
```

```python
df.groupby('animal').agg({'age':np.mean}).reset_index()
```

```python
#Count the number of each type of animal in df.
df['animal'].value_counts()
```

```python
#Sort df first by the values in the 'age' in decending order, then by the value in the 'visit' column in ascending order.
df.sort_values(["age","visits"],ascending=[False,True])
```
